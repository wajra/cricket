```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import pandas as pd
from collections import OrderedDict
import re
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url_template = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;orderby=start;page={0};spanmin1=01+Jan+1985;spanval1=span;team_view=bowl;template=results;type=team;view=innings"
```

```python
match_urls = []
for x in range(1,100):
    match_url = url_template.format(str(x))
    match_urls.append(match_url)
```

```python
bowling_performance_dict_list = []
```

```python
for url in match_urls:
    html = urllib.request.urlopen(url, context=ctx).read()
    bs = BeautifulSoup(html, "lxml")
    table_body = bs.find_all('tbody')
    rows = table_body[0].find_all('tr')
    temp_data = OrderedDict()
    for row in rows:
        cols=row.find_all('td')
        cols=[str(x.text.strip()) for x in cols]
        temp_data=OrderedDict()
        for i in range(len(cols)):
            temp_data["Team"]=cols[0]
            temp_data["Score"]=cols[1]
            temp_data["Overs"]=cols[2]
            temp_data["RPO"]=cols[3]
            temp_data["Lead"]=cols[4]
            temp_data["Inns"]=cols[5]
            temp_data["Result"]=cols[6]
            temp_data["Opposition"]=cols[8]
            temp_data["Ground"]=cols[9]
            temp_data["Date"]=cols[10]
        bowling_performance_dict_list.append(temp_data)
```

```python
bowling_df = pd.DataFrame(bowling_performance_dict_list)
```
