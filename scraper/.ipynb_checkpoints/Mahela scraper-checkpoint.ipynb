# Mahela Jayawardena

Let's write a script that downloads Mahela's batting statistics for each inning in his career.
The data comes from [espncricinfo.com](espncricinfo.com)

```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import pandas as pd
from collections import OrderedDict
import re
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
# This is the link for Mahela's innings by innings statistics
mahela_link = 'https://stats.espncricinfo.com/ci/engine/player/49289.html?class=1;template=results;type=batting;view=innings'
```

![](../images/cricinfo_shot.png)

```python
# So the plan is that we loop through each row and get the data into a Pandas dataframe
list_of_dicts = []
html = urllib.request.urlopen(mahela_link, context=ctx).read()
bs = BeautifulSoup(html, "lxml")
table_body = bs.find_all('tbody')
rows = table_body[1].find_all('tr')
temp_data = OrderedDict()
for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    temp_data=OrderedDict()
    for i in range(len(cols)):
        temp_data["Name"]='Mahela Jayawardena'
        temp_data["Country"]='Sri Lanka'
        temp_data["Runs"]=cols[0]
        temp_data["Mins"]=cols[1]
        temp_data["BF"]=cols[2]
        temp_data["4s"]=cols[3]
        temp_data["6s"]=cols[4]
        temp_data["SR"]=cols[5]
        temp_data["POS"]=cols[6]
        temp_data["Dismissal"]=cols[7]
        temp_data["Inns"]=cols[8]
        temp_data["Opposition"]=cols[10]
        temp_data["Ground"]=cols[11]
        temp_data["Date"]=cols[12]
    list_of_dicts.append(temp_data)
```

```python
# Get that into a dataframe
mahela_df = pd.DataFrame(list_of_dicts)
# Print it out
mahela_df.head(10)
```

```python
# Write this out to a csv
mahela_df.to_csv('../data/mahela_innings.csv')
```
